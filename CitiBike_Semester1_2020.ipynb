## Citi Bike Trip History 

I chose 6 months: January 2020 - June 2020 

```python
import pandas as pd
import numpy as np
import datetime
```

```python
# CSV location 
january = 'csv/JC-202001-citibike-tripdata.csv'
february = 'csv/JC-202002-citibike-tripdata.csv'
march = 'csv/JC-202003-citibike-tripdata.csv'
april = 'csv/JC-202004-citibike-tripdata.csv'
may = 'csv/JC-202005-citibike-tripdata.csv'
june = 'csv/JC-202006-citibike-tripdata.csv'
july = 'csv/JC-202007-citibike-tripdata.csv'
august = 'csv/JC-202008-citibike-tripdata.csv'
september = 'csv/JC-202009-citibike-tripdata.csv'
october = 'csv/JC-202010-citibike-tripdata.csv'
november = 'csv/JC-202011-citibike-tripdata.csv'
december = 'csv/JC-202012-citibike-tripdata.csv'
```

```python
#Reading CSVs
df_january = pd.read_csv(january, encoding="ISO-8859-1")
df_february= pd.read_csv(february, encoding="ISO-8859-1")
df_march = pd.read_csv(march, encoding="ISO-8859-1")
df_april = pd.read_csv(april, encoding="ISO-8859-1")
df_may = pd.read_csv(may, encoding="ISO-8859-1")
df_june = pd.read_csv(june, encoding="ISO-8859-1")
df_july = pd.read_csv(july, encoding="ISO-8859-1")
df_august = pd.read_csv(august, encoding="ISO-8859-1")
df_september = pd.read_csv(september, encoding="ISO-8859-1")
df_october = pd.read_csv(october, encoding="ISO-8859-1")
df_november = pd.read_csv(november, encoding="ISO-8859-1")
df_december = pd.read_csv(december, encoding="ISO-8859-1")
```

```python
#Concatenating csvs 
df_year = pd.concat([df_january, df_february, df_march, df_april, df_may, df_april, df_may, df_june,df_july,df_august,df_september,df_october,df_november,df_december])
```

```python
#Converting date columns from string to datetime 
df_year['starttime'] = df_year['starttime'].astype('datetime64[ns]')
df_year['stoptime'] = df_year['stoptime'].astype('datetime64[ns]')
```

```python
#Checking data
df_year.head()
```

```python
#Calculating age duration 
df_year['age'] = 2020-df_year['birth year']
```

```python
#Adding a gender value column 
df_year["gender_value"] = "" 
```

```python
#From the integer, indicate the gender value 
df_year.loc[df_year['gender'] == 2, 'gender_value'] = "Female"
df_year.loc[df_year['gender'] == 1, 'gender_value'] = "Male"
df_year.loc[df_year['gender'] == 0, 'gender_value'] = "Unknown"
```
